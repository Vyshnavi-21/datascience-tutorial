```python
import pandas as pd
#definig a dict containing emp data
data1={'name':['summi','harini','asin','samreen'],
       'age':[20,22,23,24],
       'address':['kerala','jammu kashmir','Allahabad','ongole'],
       'qualification':['Msc','MA','MCA','phd']}
#definig a dict containing emp data
data2={'name':['vyshu','prudhvi','riya','sana'],
       'age':[19,22,27,28],
       'address':['USA','Chennai','Allahabad','Ongole'],
       'qualification':['Btech','B.A','Bcom','phd'] }
#covert dictionary into data frame
df=pd.DataFrame(data1,index=[0,1,2,3])
#covert dictionary into data frame
df1=pd.DataFrame(data2,index=[2,3,6,7])
print(df,"\n\n",df1)
#using a .concat() method
res1=pd.concat([df,df1])
res1
#applying concat with axes
#join="inner" we get intersection of dataframe
res2=pd.concat([df,df1],axis=1,join='inner')
res2
```

```python
import pandas as pd
#definig a dict containing emp data
data1={'name':['summi','harini','asin','samreen'],
       'age':[20,22,23,24],
       'address':['kerala','jammu kashmir','Allahabad','ongole'],
       'qualification':['Msc','MA','MCA','phd']}
#definig a dict containing emp data
data2={'name':['vyshu','prudhvi','riya','sana'],
       'age':[19,22,27,28],
       'address':['USA','Chennai','Allahabad','Ongole'],
       'qualification':['Btech','B.A','Bcom','phd'] }
#covert dictionary into data frame
df=pd.DataFrame(data1,index=[0,1,2,3])
#covert dictionary into data frame
df1=pd.DataFrame(data2,index=[2,3,6,7])
print(df,"\n\n",df1)
#using ignore_index
res=pd.concat([df,df1],ignore_index=True)
res
```

```python
import pandas as pd
#definig a dict containing emp data
data1={'name':['summi','harini','asin','samreen'],
       age':[20,22,23,24]',
       'address':['kerala','jammu kashmir','Allahabad','ongole'],
       'qualification':['Msc','MA','MCA','phd']}
#covert dictionary into data frame
df=pd.DataFrame(data1,index=[0,1,2,3])
s1=pd.Series([1000,2000,3000,4000],name='salary')
print(df,"\n\n",s1)
res=pd.concat([df,s1],axis=1)
res
```

```python
#merging 
import pandas as pd
#define a dictionary containing emp data
data1={'key':['K0','K1','K2','K3'],
       'name':['Summi','Harini','Rani','riya'],
       'age':[20,22,23,24]}
data2={'key':['K0','K1','K2','K3'],
        'address':['kerala','jammu kashmir','Allahabad','ongole'],
       'qualification':['Msc','MA','MCA','phd']}
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,"\n\n",df1)
#using .merge() function
res=pd.merge(df,df1,on='key')
res
        
```

```python
import pandas as pd
#define a dictionary containing emp data
data1={'key':['K0','K1','K2','K3'],
       'key1':['K0','K1','K0','K1'],
       'name':['Summi','Harini','Rani','riya'],
       'age':[20,22,23,24]}
data2={'key':['K0','K1','K2','K3'],
       'key1':['K0','K0','K0','K0'],
       'address':['kerala','jammu kashmir','Allahabad','ongole'],
       'qualification':['Msc','MA','MCA','phd']}
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,"\n\n",df1)
#using .merge() function
res1=pd.merge(df,df1,on=['key','key1'])
res1
```
