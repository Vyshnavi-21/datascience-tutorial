```python
#stacked bar graph
import matplotlib.pyplot as plt
categories=['transport','income tax','housing']
values1=[15,24,30]
values2=[20,18,25]
plt.bar(categories,values1,label='first half ',color='purple')
plt.bar(categories,values2,bottom=values1,label='second half ',color='orange')
plt.xlabel('categories')
plt.ylabel('VALUES')
plt.title('stacked bar graph')
plt.legend()
plt.show()
```

```python
#histogram-no of times an item appear repeatedly
import matplotlib.pyplot as plt
x=[70,80,89,80,70,40,50,50,100,70,80,80,30,70,100,80]
plt.hist(x,orientation="vertical",color='pink')
plt.title('Mark Analysis')
plt.ylabel('No of Students')
plt.xlabel('marks')
plt.show()

```

```python
#sstacked histogram with multiple datasets
import matplotlib.pyplot as plt
import numpy as np
mark1=np.array([2,4,7,6,8,9,10,12,14,13,25,16])
mark2=np.array([3,9,7,4,5,16,10,13,24,11,15,28])
plt.hist([mark1,mark2],bins=10,stacked=True,color=['skyblue','salmon'],edgecolor='black')
plt.title('stacked histogram with multiple dataset')
plt.label('count')
plt.xlabel('marks')
plt.legend(['mark Dataset1','mark dataset2'])
plt.show()
```
